```python
import networkx as nx
import matplotlib.pyplot as plt
g1 = nx.petersen_graph()
nx.draw(g1, with_labels=True)
plt.show()
```

```python
G = nx.Graph()
```

```python
G
```

```python
class Location:
    def __init__(self, name, description):
        self.name = name
        self.description = description
    
    def __str__(self):
        return self.name
```

```python
G.add_nodes_from(
    [("Johns Mansion", Location("Johns Mansion", "It is the place where John lives),
     (),
     (),
     (),
     (),
     (),
     (),
     ()]
)
```

```python
kitchen = Location("Kitchen", "Place where all the cooking stuff is located")
living_room = Location("Living Room", "Here the Lin family is hanging out most of the time")
restroom = Location("Restroom", "Place to wash the hands, take a shower and go to the toilet")
```

```python
johns_flat = nx.Graph()
johns_flat.add_nodes_from(
    [(kitchen.name, {"location": kitchen}),
     (living_room.name, {"location": living_room}),
     (restroom.name, {"location": restroom})]
)

johns_flat.add_edge(kitchen.name, living_room.name)
johns_flat.add_edge(restroom.name, living_room.name)
```

```python
johns_flat.adj
```

```python
nx.draw(johns_flat, with_labels=True)
plt.show()
```

```python
list(johns_flat.nodes)
```
