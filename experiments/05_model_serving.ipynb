```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
```

```python
model_checkpoint = "HuggingFaceH4/zephyr-7b-alpha"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForCausalLM.from_pretrained(model_checkpoint, device_map="auto", torch_dtype=torch.bfloat16)
```

```python
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    do_sample=True,
    max_new_tokens=500,
    temperature=0.3,
    top_p=0.95,
    top_k=155,
    repetition_penalty=1.15
)
```

```python
pipe("What is concious thinking?")
```

```python
from tqdm import tqdm
from torch.utils.data import Dataset

class MyDataset(Dataset):
    x = ["What is concious thinking?",
     "How do you know if you are concious?",
     "What is reality?", 
     "When will the world end?",
     "Why is the sky blue?",
     "When is the next world war?",
     "What is a black hole?",
     "What is a quark?",
     "What is a photon?",
     "What is a gluon?"
     "Is there a god?",
     "What is the meaning of life?",
     "What is the meaning of death?",
     "What is the meaning of conciousness?",
     "What is the meaning of reality?",
     "What is the meaning of existence?",
     "What is the meaning of the universe?",
     "What is the meaning of the multiverse?",
     "When does the universe end?",
     "What is the universe expanding into?"]

    def __len__(self):
        return len(self.x)

    def __getitem__(self, i):
        return self.x[i]

dataset = MyDataset()

responses = dict()

for batch_size in [1, 5, 10, 20]:
    responses[batch_size] = []
    print("-" * 30)
    print(f"Streaming batch_size={batch_size}")
    for out in tqdm(pipe(dataset, batch_size=batch_size), total=len(dataset)):
        responses[batch_size] += out
```

```python
for response in responses[1]:
    print("*"*30)
    print(response["generated_text"])
```
